* ¿Cual fue el rango de edad mas vacunado hasta la fecha?

* ¿Que vacuna fue la mas aplicada?
* ¿Que lugar, donde se aplicaron las vacunas, fue el mas utilizado?

###### Abril fue el mes donde mas dosis se le aplicaron a los civiles, mientras que Enero fue donde se dieron menos dosis.
###### Las personas de 51 a 60 años fueron las personas mas vacunas hasta la fecha
###### Las mujeres fueron mas vacunadas a comparacion de los hombres.
###### Como podemos observar la vacuna Sputnik fue la mas usada con un 47,4%, mientras la menos utilizada fue la moderna con un 2,2%.
###### De los 3 tipos de lugares que hay, para vacunar a las personas, los lugares publicos fueron los que mas se usaron.
___

### Conclusion
La conclusion que podemos obtener en base a las preguntas y las respuestas obtenidas son que las personas entre las edades de 51 a 60 años fueron las mas vacundas ya que esto se debera a que son de factor de riesgo. El mes donde mas dosis se aplicaron en las personas fue Enero (en lugares publicos), luego tenemos que priorizaron mas la vacunacion en las mujeres que en los hombres y que las vacunas mas utilizadas fueron las de Sputnik.
