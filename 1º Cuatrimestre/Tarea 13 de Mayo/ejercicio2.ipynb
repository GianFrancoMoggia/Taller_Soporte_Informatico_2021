```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df1=pd.read_csv('beneficiarios_primera_casa_BA-2018.csv')

y_values_edad = df1['edad'].value_counts()
x_values_edad= y_values_edad.keys()
y_values_edad = y_values_edad.values

#grafico1
plt1.bar(x_values_edad,y_values_edad)
plt1.title('CUANTAS EDADES HAY DE CADA UNA')
plt1.xlabel('PERSONAS DE DIFERENTES EDADES')
plt1.ylabel('CANTIDAD DE CADA EDAD')
plt1.show()

#____________________________________________________________________________
y_values_alquila = df1['alquila'].value_counts()
x_values_alquila= y_values_alquila.keys()
y_values_alquila = y_values_alquila.values

#grafico2
plt2.bar(x_values_alquila,y_values_alquila)
plt2.title('CUANTAS PERSONAS ALQUILAN')
plt2.xlabel('ALQUILA')
plt2.ylabel('CANTIDAD QUE ALQUILA')
plt2.show()

'''
1-  ¿Qué conclusión o conclusiones se atrevería a sacar del gráfico obtenido?
2-  ¿Cuáles serían las variables que considera que deberían tenerse en
    cuenta pero no las puede apreciar porque el csv no se las provee?
    ¿Porqué?
3-  ¿Podría relacionar los datos de las dos columnas elegidas para especular
    y proponer una conclusión?
    En caso de que no, ¿Con qué columnas cree que lo podría hacer? ¿Porqué?

1-Se le pregunto a una gran variedad de personas de distintas edades y la mayoria alquila
2-Ninguna, ya que este csv esta completo con respecto a los datos que te provee
3-Si (Sacar un promedio de las personas q alquilan segun la edad)
'''
```
